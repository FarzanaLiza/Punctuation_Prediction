## Our covered punctuations

```python
special_labels = {',': 'I-COMMA',
                  '.': 'I-DOT',
                  '?': 'I-QMARK',
                  '!': 'I-EMARK',
                  ':': 'I-COLON',
                  ';': 'I-SEMICOLON',
                  '-': 'I-HYPHEN', # Label for the hyphen (new)
                  "'": 'I-APOST'  } # Label for the apostrophe (new)
normal_label = 'O'
```

```python
labels_set = list(special_labels.values()) + [normal_label]
```

```python
from simpletransformers.ner import NERModel, NERArgs
import pandas as pd
```

```python
model_args = NERArgs()
model_args.overwrite_output_dir = True
model_args.save_steps = -1
model = NERModel('bert',
                 './bert_models/8/checkpoint-866-epoch-1/',
                 labels = labels_set,
                 args=model_args,
                 use_cuda=False)
```

### Split the text based on 500 word counts
If the paragraph is too big, we need to split the text

```python
words = YOUTUBE_TEXTS[0].split()  # Split the single string into words
word_groups = [words[i:i+510] for i in range(0, len(words), 510)]  # Create groups of 500 words

# Join each group of words into a string and create a new list
new_paragraph = [" ".join(group) for group in word_groups]

#print(new_paragraph)
```

### Test on youtube paragraph

```python
removed_punctuations = []
for txt in new_paragraph:
    txt2 = txt
    for pun in special_labels.keys():
        txt2 = txt2.replace(pun, "")
    removed_punctuations.append(txt2)
```

```python
#removed_punctuations
```

* Use "model.prediction" when it is undupervised data.
*  Use "model.eval_model" when it is supervised data that is label is already created

```python
result, _ = model.predict(removed_punctuations)
```

```python
inv_special_labels = {v: k for k, v in special_labels.items()}

predicted_texts = []

for res in result:
    sentence = []
    for diction in res:
        sentence.append(list(diction.keys())[0])
        if list(diction.values())[0] in list(inv_special_labels.keys()):
            sentence.append(inv_special_labels[list(diction.values())[0]])
    predicted_texts.append(" ".join(sentence))
```
